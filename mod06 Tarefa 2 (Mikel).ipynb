```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook
```

```python
sns.set_theme()
sns.set_style('dark')
```

```python
sinasc_RO = pd.read_csv('sinasc_RO_2019.csv')
sinasc_RO.head(2)
```

```python
sinasc_RO.columns
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.boxplot(data = sinasc_RO['IDADEMAE'], color = 'yellow', linecolor = 'blue')
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(data = sinasc_RO['PESO'], color = 'pink', linecolor = 'green')
```

####  3) Faça um histograma da idade da mãe

```python
sns.histplot(data = sinasc_RO['IDADEMAE'], bins = 10, color = 'brown', label='Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência') 
plt.legend(loc='best')
```

#### 4) Faça um histograma do peso do bebê

```python
sns.histplot(data = sinasc_RO['PESO'], bins = 10, color = 'blue', label='Peso do Bêbê')
plt.xlabel('Peso do Bêbê')
plt.ylabel('Frequência') 
plt.legend(loc='best')
```
