```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
sinasc_RO = pd.read_csv('sinasc_RO_2019.csv')
sinasc_RO.head(2)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
sinasc_filtrado = sinasc_RO[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]
sinasc_filtrado.head()
```

```python
sinasc_filtrado['GESTACAO'].unique()
```

```python
plt.rc('figure', figsize=(20, 6))

# Criando os subplots
fig, axes = plt.subplots(2, 3)

plt.subplots_adjust(top=0.9, hspace=0.3, wspace=0.5)

# Criando count plots
sns.countplot(ax=axes[0, 0], x='APGAR1', data=sinasc_filtrado)
sns.countplot(ax=axes[0, 1], x='APGAR5', data=sinasc_filtrado)
sns.countplot(ax=axes[1, 0], x='SEXO', data=sinasc_filtrado)
sns.countplot(ax=axes[0, 2], x='GESTACAO', data=sinasc_filtrado)
sns.countplot(ax=axes[1, 1], x='ESTCIVMAE', data=sinasc_filtrado)

#Renomeando os labels
axes[0, 0].set_xlabel("APGAR 1-minuto") 
axes[0, 1].set_xlabel("APGAR 5-minuto")  
axes[1, 0].set_xlabel("Sexo do Bêbê") 
axes[0, 2].set_xlabel("Gestação") 
axes[1, 1].set_xlabel("Estado Civil da Mãe") 


# Rotate x-tick labels and change their color
axes[0, 0].set_xticklabels(axes[0, 0].get_xticklabels(), color='green')
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), color='green')
axes[1, 0].set_xticklabels(axes[1, 0].get_xticklabels(), color='green')
axes[0, 2].set_xticklabels(axes[0, 2].get_xticklabels(), rotation=45, ha='right', color='green')
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), color='green')

# Turn off the last subplot
axes[1, 2].axis('off')

# Show the plot
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(y = 'CONSULTAS', x = 'APGAR5', estimator= 'mean', data= sinasc_RO )
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize = (20,10))
plt.subplots_adjust(top=0.8, hspace= 0.5, wspace= 1.2, bottom= 0.7)
fig, axes = plt.subplots(2,2)

sns.countplot(ax = axes[0, 0], x = 'SEXO', data = sinasc_RO)
sns.countplot(ax = axes[0, 1], x = 'GESTACAO', data = sinasc_RO)
sns.countplot(ax = axes[1, 0], x = 'ESTCIVMAE', data = sinasc_RO)
sns.countplot(ax = axes[1, 1], x = 'CONSULTAS', data = sinasc_RO)

# Mudando a cor dos labels
axes[0, 0].set_xticklabels(axes[0, 0].get_xticklabels(), color='green')
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=20, ha='right', color='brown')
axes[1, 0].set_xticklabels(axes[1, 0].get_xticklabels(), color='purple')
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), color='brown')
```
